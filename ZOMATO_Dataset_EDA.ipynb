```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
final_df['Country'].value_counts()
```

```python
country_names = final_df['Country'].value_counts().index
```

```python
country_values = final_df['Country'].value_counts().values
```

```python
##Pie chart - Top 3 countries that uses Zomato
plt.pie(country_values[:3],labels=country_names[:3],autopct="%1.1f%%");
```

**Observation: Zomato maximum records and transaction are from India,After that USA and then UK**

```python
final_df.columns
```

```python
ratings =final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

**Observation:**
1. **Whenever Rating is between 4.5 to 4.9 --> Excellent**
2. **Whenever Rating is between 4.0 to 4.4 --> Very Good**
3. **Whenever Rating is between 3.5 to 3.9 --> Good**
4. **Whenever Rating is between 2.5 to 3.4 --> Average**
5. **Whenever Rating is between 1.8 to 2.4 --> Poor**

```python
ratings.head(2)
```

```python
plt.figure(figsize=(12, 4))
custom_palette = sns.color_palette("viridis")
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings,palette=custom_palette)
```

```python
plt.figure(figsize=(12, 4))
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings,hue='Rating color',palette=['lightblue','red','orange','yellow','green','darkgreen'])
```

**Observation:**
1. **Not Rated count is very high**
2. **Maximum Number of ratings are between 3.0 to 3.6**

```python
#count plot
plt.figure(figsize=(8,5))
sns.countplot(x='Rating color',palette=['lightblue','red','orange','yellow','green','darkgreen'],data=ratings)
```
